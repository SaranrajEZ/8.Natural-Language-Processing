```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('news.csv')
dataset = df.drop("Unnamed: 0", axis=1)

dataset
```

```python
y = dataset["label"]
X_train, X_test, y_train, y_test = train_test_split(dataset['text'], y, test_size=0.33, random_state=53)
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)

print(tfidf_train)
```

```python
print(tfidf_train.toarray())
```

```python
len(tfidf_vectorizer.get_feature_names_out())
```

```python
clf = MultinomialNB()
clf.fit(tfidf_train, y_train)
```

```python
pred = clf.predict(tfidf_test)
score = metrics.accuracy_score(y_test, pred)
print("Accuracy: %0.3f" % score)
```

```python
cm = metrics.confusion_matrix(y_test, pred, labels=['FAKE', 'REAL'])
print("Confusion Matrix:\n", cm)

report = metrics.classification_report(y_test, pred)
print("Classification Report:\n", report)
```

```python
sample_index = 4
sample_text = X_train.iloc[[sample_index]]
sample_transformed = tfidf_vectorizer.transform(sample_text)
print("Predicted label for the first sample:", clf.predict(sample_transformed))
```
